## 4. Explore the data

```python
Pisa=pd.read_csv("pisaresearch.csv")
```

```python
Pisa.info()
```

```python
Pisa.head()
```

```python
Pisa_freq=pd.melt(Pisa,id_vars=['Country'],var_name="subject",value_name="frequence")
```

```python
Pisa_freq.head()
```

```python
Pisa_freq1=Pisa_freq.groupby(['Country','subject','frequence'])['frequence'].count()
Pisa_freq1.head(20)
```

```python
Pisa_new=Pisa_freq1.unstack()
```

```python
neworder=["Always or almost always","Often","Sometimes","Never or rarely"]
Pisa_new=Pisa_new.reindex(columns=neworder)
Pisa_new.head()
```

```python
Pisa_new['always_percentage']=round(Pisa_new["Always or almost always"]/(Pisa_new["Always or almost always"]+Pisa_new["Never or rarely"]+Pisa_new["Often"]+Pisa_new["Sometimes"]),2)
```

```python
Pisa_new['Often_percentage']=round(Pisa_new["Often"]/(Pisa_new["Always or almost always"]+Pisa_new["Never or rarely"]+Pisa_new["Often"]+Pisa_new["Sometimes"]),2)
```

```python
Pisa_new['Sometimes_percentage']=round(Pisa_new["Sometimes"]/(Pisa_new["Always or almost always"]+Pisa_new["Never or rarely"]+Pisa_new["Often"]+Pisa_new["Sometimes"]),2)
```

```python
Pisa_new['Rarely_percentage']=round(Pisa_new["Never or rarely"]/(Pisa_new["Always or almost always"]+Pisa_new["Never or rarely"]+Pisa_new["Often"]+Pisa_new["Sometimes"]),2)
```

```python
Pisa_research=Pisa_new[['always_percentage',"Often_percentage","Sometimes_percentage","Rarely_percentage"]]
```

```python
Pisa_research=pd.DataFrame(Pisa_research.stack())
Pisa_research.head()
```

```python
Pisa_research.to_csv("Pisa_freq.csv")
```
